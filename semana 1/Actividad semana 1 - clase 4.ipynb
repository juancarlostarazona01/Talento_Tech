```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a dataset
# Let´s build the values lists
nombre = ['Pedro', 'Pablo', 'Marcelo', 'Alberto', 'Jose',
          'Maria', 'Rosana', 'Julia', 'Luisa', 'Gabriela']
print(nombre)

promedio_Lenguaje    = [3.8, 3.6, 4.0, 3.9, 4.1, 4.3, 4.0, 4.0, 3.7, 3.0]
promedio_Ingles      = [4.1, 4.5, 3.8, 3.3, 4.0, 2.0, 3.7, 2.5, 4.5, 2.5]
promedio_Matematicas = [3.0, 3.2, 2.8, 3.2, 4.4, 4.4, 3.3, 3.0, 4.5, 3.2]
promedio_artes       = [4.0, 4.0, 4.0, 4.0, 4.5, 4.5, 4.0, 3.5, 3.0, 4.0]
promedio_Quimica     = [5.0, 3.5, 3.5, 4.0, 3.5, 3.0, 4.5, 3.5, 3.0, 5.0]
promedio_Fisica      = [3.4, 3.2, 3.3, 3.1, 3.8, 3.0, 3.0, 2.9, 4.5, 3.5]
necesario_aprobado   = [3.2, 3.3, 3.1, 3.1, 2.2, 3.2, 3.7, 3.5, 2.8, 2.2]
# value_bmi_list will be calculated: bmi = w/(h^2)
#value_bmi_list = [25,25,25] # init values [kg/m2]

# Let´s use a dictionary ({key:value})
# with the format {"str":list}
data_dict = {"nombre":nombre,
             "promedio_lenguaje":promedio_Lenguaje,
             "promedio_ingles":promedio_Ingles,
             "promedio_matematicas":promedio_Matematicas,
             "promedio_artes":promedio_artes,
             "promedio_quimica":promedio_Quimica,
             "promedio_fisica":promedio_Fisica,
             "necesario_aprobado":necesario_aprobado} #,
              #"BMI":value_bmi_list}

print(data_dict)

# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)

print(df)
# Call a variable
df
print(df)
df
```

```python
# prompt: Con DataFrame df: add another column

df['promedio_ano'] = (df['promedio_lenguaje'] + df['promedio_ingles'] + df['promedio_matematicas']
                 + df['promedio_artes'] + df['promedio_quimica'] + df['promedio_fisica']) / 6
df
```

```python
df["paso_materia"] = np.where((df['promedio_ano'] >= df['necesario_aprobado']), 'pasó el año :)', 'no pasó el año :(')
df
```

```python
df['promedio_ano'].mean()
```

```python
import matplotlib.pyplot as plt

df['promedio_ano'].plot(style='.-', kind='line', figsize=(8, 4), title='promedios año', markersize=10)
plt.gca().spines[['top', 'right']].set_visible(False)
plt.axhline(3.66, linewidth = 2, linestyle = "dashed")
plt.title("3.66 nota promedio curso")
plt.grid()
```
