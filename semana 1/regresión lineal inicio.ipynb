```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 2, 3, 4, 4, 5, 6]
y = [2, 3, 4, 4, 4, 6, 5, 7]

data_dict = {"x":x, "y":y}
print(data_dict)

df = pd.DataFrame(data_dict)
```

```python
df['xy'] = df['x'] * df['y']
df['x^2'] = df['x'] ** 2
df
```

```python
sum_x = df['x'].sum()
sum_y = df['y'].sum()
sum_xy = df['xy'].sum()
sum_x2 = df['x^2'].sum()
print(sum_x, sum_y, sum_xy, sum_x2)
```

```python
m = (len(df) * sum_xy - sum_x * sum_y) / (len(df) * sum_x2 - sum_x * sum_x)
m
```

```python
b = (sum_y - m * sum_x) / len(df)
b
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='x', y='y', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.plot(df['x'], m * df['x'] + b, color='red')
plt.show()
```
