```python
x = [1, 2, 3, 4, 5]
h = [-1, 5, 3, -2, 1]
```

```python
# Hecho enm chatGPT con el prompt "como aplico convolución de forma manual para un x y un h de 5 datos cada uno"

def convolucion_manual(x, h):
    N = len(x)
    M = len(h)

    # Tamaño de la señal resultante
    tamaño_salida = N + M - 1

    # Inicializar la salida con ceros
    salida = [0] * tamaño_salida

    # Calcular la convolución
    for n in range(tamaño_salida):
        for k in range(N):
            if 0 <= n - k < M:
                salida[n] += x[k] * h[n - k]

    return salida

# Aplicar la convolución
resultado = convolucion_manual(x, h)
print(resultado)
```
