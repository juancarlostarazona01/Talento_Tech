```python
import keras
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# Generación de datos de clasificación sintéticos
X, y = make_classification(n_samples=100, n_features=5, n_classes=2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Conversión de etiquetas a codificación one-hot, se establece que el número de clases tiene que ser de 2
y_train_categorical = keras.utils.to_categorical(y_train, num_classes=2)
y_test_categorical = keras.utils.to_categorical(y_test, num_classes=2)
```

```python
# Creación de una instancia de un modelo secuencial de Keras
model = keras.models.Sequential()
```

```python
# Capa oculta con 10 neuronas y activación ReLU
model.add(keras.layers.Dense(10, input_shape=(X_train.shape[1],), activation='relu'))

# Capa de salida con 2 neuronas (para cada clase) y activación softmax
model.add(keras.layers.Dense(2, activation='softmax'))
```

```python
# Compilación
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Entrenamiento
model.fit(X_train, y_train_categorical, epochs=10, batch_size=32, validation_data=(X_test, y_test_categorical))
```
