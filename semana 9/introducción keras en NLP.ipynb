```python
import keras
import tensorflow as tf
```

```python
frases = [
    'Hola mundo',
    'Hola a todos',
    'Hola a todo el mundo'
]
```

```python
# Generaci贸n del diccionario de tokens
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words = 10)
tokenizer.fit_on_texts(frases)
word_index = tokenizer.word_index
print('word_index =', word_index)
```

```python
# Generaci贸n de una secuencia tokenizada
secuencias = tokenizer.texts_to_sequences(frases)
print('secuencias =', secuencias)
```

```python
# Relleno a longitud uniforme
relleno = tf.keras.preprocessing.sequence.pad_sequences(secuencias)
print('relleno =\n', relleno)
```

```python
frases = [
    'Hola mundo',
    'Hola a todos',
    'Hola a todo el mundo',
    'Buen dia, como estas hoy']

# Generaci贸n del diccionario de tokens
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words = 10, oov_token="<00V>")
tokenizer.fit_on_texts(frases)
word_index = tokenizer.word_index
print('\nword_index =', word_index)

# Generaci贸n de una secuencia tokenizada
secuencias = tokenizer.texts_to_sequences(frases)
print('secuencias =', secuencias)

# Relleno a longitud uniforme
relleno = tf.keras.preprocessing.sequence.pad_sequences(secuencias)
print('relleno =\n', relleno)
```

```python
# Relleno a longitud uniforme
relleno2 = tf.keras.preprocessing.sequence.pad_sequences(secuencias,
                                                        padding = 'post',
                                                        truncating = 'post')

print('relleno =\n', relleno2)
```
