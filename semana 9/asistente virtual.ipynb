```python
from datetime import datetime

def pedir_cita():
    while True:
        fecha_input = input("Ingresa la fecha de la cita (formato: YYYY-MM-DD): ")
        hora_input = input("Ingresa la hora de la cita (formato: HH:MM): ")

        try:
            # Combina la fecha y la hora en un solo objeto datetime
            fecha_cita = datetime.strptime(f"{fecha_input} {hora_input}", "%Y-%m-%d %H:%M")

            # Verifica si la fecha y hora es en el futuro
            if fecha_cita > datetime.now():
                print(f"Cita programada para el {fecha_cita.strftime('%Y-%m-%d a las %H:%M')}")
                break
            else:
                print("La fecha y hora ingresadas ya pasaron. Intenta con un horario futuro.")
        except ValueError:
            print("Fecha u hora no válidas. Por favor, intenta de nuevo con el formato correcto.")
    confirmacion()
```
