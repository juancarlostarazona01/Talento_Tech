```python
import numpy as np
from math import sqrt
#from math import abs

x1 = int(input('x1: '))
y1 = int(input('y1: '))
x2 = int(input('x2: '))
y2 = int(input('y2: '))
d = sqrt((x2-x1)**2+(y2-y1)**2)
print("distancia euclidiana", d)
d2 = abs(x2-x1)+abs(y2-y1)
print("distancia de manhattan", d2)
d3 = max(abs(x2-x1),abs(y2-y1))
print("distancia de chebyshev", d3)
```

```python
x = np.array([ 1, 2, 3, 4 ])
y = np.array([ 100, -28, 3456, 0.1 ])
canberra = np.sum(np.abs(x-y))/np.sum(np.abs(x)+np.abs(y))
print("distancia de canberra", canberra)
```

```python
import numpy as np
from numpy.linalg import norm

coseno = np.dot(x,y)/(norm(x)*norm(y))
print("coseno", coseno)
```

```python
palabra1 = input('palabra1: ')
palabra2 = input('palabra2: ')
hamming = 0
for i in range(len(palabra1)):
    if palabra1[i] != palabra2[i]:
        hamming += 1
print(hamming)
d5 = 1-(hamming/max(len(palabra1),len(palabra2)))
print("distancia de ", d5)
```
