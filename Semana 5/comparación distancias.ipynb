```python
import pandas as pd

# Read the iris dataset, handling the concatenated 'Species' column
iris = pd.read_csv("iris.data", names=["SepalLengthCm",
                                                      "SepalWidthCm",
                                                      "PetalLengthCm",
                                                      "PetalWidthCm",
                                                      "Species"])

# Split the concatenated 'Species' column into individual rows
iris = iris.assign(Species=iris['Species'].str.findall(r'Iris-[a-z]+')).explode('Species')

# Calculate the mean of numerical columns for each species
iris.groupby('Species').mean(numeric_only=True)
```

```python
setosa_sepal_length = iris[iris['Species'] == 'Iris-setosa']['SepalLengthCm']
x1 = setosa_sepal_length.mean()
print(x1)
setosa_petal_length = iris[iris['Species'] == 'Iris-setosa']['SepalWidthCm']
x2 = setosa_petal_length.mean()
print(x2)
setosa_sepal_length1 = iris[iris['Species'] == 'Iris-versicolor']['SepalLengthCm']
y1 = setosa_sepal_length1.mean()
print(y1)
setosa_petal_length1 = iris[iris['Species'] == 'Iris-versicolor']['SepalWidthCm']
y2 = setosa_petal_length1.mean()
print(y2)
setosa_sepal_length2 = iris[iris['Species'] == 'Iris-virginica']['SepalLengthCm']
z1 = setosa_sepal_length2.mean()
print(z1)
setosa_petal_length2 = iris[iris['Species'] == 'Iris-virginica']['SepalWidthCm']
z2 = setosa_petal_length2.mean()
print(z2)
```

```python
from math import sqrt

distanciasetvereuc = sqrt((x2-x1)**2+(y2-y1)**2)
print("distancia setosa sepalo euclidiana", distanciasetvereuc)
distanciasetverman = abs(x2-x1)+abs(y2-y1)
print("distancia setosa sepalo de manhattan", distanciasetverman)
distanciasetvireuc = sqrt((x2-x1)**2+(z2-z1)**2)
print("distancia verticolor sepalo euclidiana", distanciasetvireuc)
distanciasetvirman = abs(x2-x1)+abs(z2-z1)
print("distancia verticolor sepalo de manhattan", distanciasetvirman)
distanciavervireuc = sqrt((y2-y1)**2+(z2-z1)**2)
print("distancia virginica sepalo euclidiana", distanciavervireuc)
distanciavervirman = abs(y2-y1)+abs(z2-z1)
print("distancia virginica sepalo de manhattan", distanciavervirman)
```

```python
import matplotlib.pyplot as plt
plt.scatter([x1, x2], [y1, y2])
plt.plot([x1, x2], [y1, y2], label='Euclidean')  # Line for Euclidean distance

# For Manhattan, draw a horizontal then vertical line
plt.plot([x1, x2], [y1, y1], color='red', linestyle='--', label='Manhattan')
plt.plot([x2, x2], [y1, y2], color='red', linestyle='--')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Distance Visualization')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter([x1, x2], [z1, z2])
plt.plot([x1, x2], [z1, z2], label='Euclidean')  # Line for Euclidean distance

# For Manhattan, draw a horizontal then vertical line
plt.plot([x1, x2], [z1, z1], color='red', linestyle='--', label='Manhattan')
plt.plot([x2, x2], [z1, z2], color='red', linestyle='--')

plt.legend()
plt.xlabel('X')
plt.ylabel('Z')
plt.title('Distance Visualization')
plt.show()
```
